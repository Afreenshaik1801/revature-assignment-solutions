```python
#Importing the requried Moduels
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

get_ipython().run_line_magic('matplotlib', 'inline')

sns.set(style='whitegrid')
```

```python
n = 1000 #no of records

dates = pd.date_range(start='2022-01-01', end='2024-12-31').to_pydatetime().tolist()
product_categories = ['Machinery', 'Tools', 'Electrical', 'Accessories']
regions = ['North','South','East','West']
customer_types = ['Corporate','Distributor','Retail']
sales_channels = ['Online','Offline']

data = {
    'Order_ID': np.arange(1001, 1001+n),
    'Date': np.random.choice(dates, n),
    'Product_Category': np.random.choice(product_categories, n),
    'Product_Name': np.random.choice(['Pump', 'Compressor', 'Welding Machine', 'Conveyor Belt', 
                                      'Drill', 'Motor', 'Valve', 'Gearbox'], n),
    'Region': np.random.choice(regions, n),
    'Customer_Type': np.random.choice(customer_types, n),
    'Units_Sold': np.random.randint(10, 200, n),
    'Unit_Price': np.random.randint(500, 5000, n),
    'Discount': np.random.uniform(0, 0.25, n),
    'Sales_Channel': np.random.choice(sales_channels, n)
}
df = pd.DataFrame(data)

df['Total_Sales'] = df['Units_Sold'] * df['Unit_Price'] * (1 -df['Discount'])
df['Profit'] = df['Total_Sales'] * np.random.uniform(0.05,0.2,n)
df
```


                                                Univarient Analysis

```python
#Distribution of Total Sales in Company
sns.histplot(x='Total_Sales',data=df,bins=10,kde = True)
plt.title("Distribution of Total Sales")
plt.xlabel("Total Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot of high-profit outliers
plt.figure(figsize=(10,6))
sns.boxplot(x=df['Profit'], color='lightblue')
plt.title('Profit Boxplot')
plt.xlabel('Profit')
plt.show()
```

```python
#countplot between Machinery and Electrical
df_filtered = df[df['Product_Category'].isin(['Machinery','Electrical'])]
plt.figure(figsize=(8,6))
ax = sns.countplot(x = 'Product_Category',data = df_filtered)
ax.bar_label(ax.containers[0])
plt.title('Count of orders by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Number of Orders')
plt.show()
```

```python
#Countplot for Orders in a region
df_filtered = df[df['Region'].isin(['South','West'])]
plt.figure(figsize=(8,6))
ax = sns.countplot(x='Region',data=df_filtered)
ax.bar_label(ax.containers[0])
plt.title('Number of Orders by Region')
plt.xlabel('Region')
plt.ylabel('Number of Orders')
plt.show()
```

```python
#Count Plot between the sales Channel
ax = sns.countplot(x = 'Sales_Channel',data = df)
ax.bar_label(ax.containers[0])
plt.title("Number of Orders by Sales Channels")
plt.xlabel("Sales Channel")
plt.ylabel("Number of Orders")
plt.show()
```

```python
# Analysis between region and Total_Sales
sns.barplot(x='Region',y='Total_Sales',data=df)
plt.title("Region vs Total Sales")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Analysis between sales Channel vs Customer Type
plt.figure(figsize=(8,6))
sns.countplot(x='Sales_Channel', hue='Customer_Type', data=df, palette='pastel')
plt.title('Sales Channel vs Customer Type')
plt.xlabel('Sales Channel')
plt.ylabel('Number of Orders')
plt.show()
```

```python
#Matrix Between Total Sales and Profit
corr = df[['Total_Sales','Profit','Discount','Units_Sold','Total_Sales','Unit_Price']].corr()
sns.heatmap(corr,annot=True)
plt.title("Correlation Matrix")
plt.show()
```

                            #Finding the Tread Analysis

```python
#Analyzing the steady growth in sales from 2022 to 2024

#Yearly Sales Trend
yearly_sales = df.groupby(df['Date'].dt.year)['Total_Sales'].sum()
plt.figure(figsize=(8,5))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o')
plt.title('Yearly Sales Trend (2022-2024)')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(yearly_sales.index)
plt.show()
```

```python
#Analysis of Monthly sales
monthly_sales = df.groupby(df['Date'].dt.month)['Total_Sales'].sum()
plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title('Monthly Sales Trend (Combined for 2022-2024)')
plt.xlabel('Month (1 = Jan ... 12 = Dec)')
plt.ylabel('Total Sales')
plt.xticks(range(1,13))
plt.show()
```

```python
df['Year'] = df['Date'].dt.year
df['Quarter'] = df['Date'].dt.to_period('Q').astype(str)

quarterly_sales = df.groupby(['Quarter', 'Year'])['Total_Sales'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Quarter', y='Total_Sales', data=quarterly_sales, hue='Year', palette='viridis')

plt.title('Quarterly Sales Trend', fontsize=16)
plt.xlabel('Quarter', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Category Performance in a yearly manner
plt.figure(figsize=(10,6))
sns.countplot(x='Year',hue='Product_Category', data=df)
plt.title("Product Category slales Per Year")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```
