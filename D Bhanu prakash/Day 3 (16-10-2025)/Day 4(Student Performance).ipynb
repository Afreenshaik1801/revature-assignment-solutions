**To performe analysis of a student with synthetic data set using pandas and matplotlib/seaborn**

```python
import pandas as pd
import numpy as np
import seaborn as se
import matplotlib.pyplot as plt
se.set(style="whitegrid")
pd.set_option('display.max_columns', None)
```

```python
num_students=100
data = {
    'Student_ID': range(1, num_students + 1),
    'Gender': np.random.choice(['Male', 'Female'], size=num_students),
    'Age': np.random.randint(14, 20, size=num_students),
    'Study_Hours': np.round(np.random.normal(3.5, 1.2, num_students), 1),
    'Attendance_%': np.round(np.random.normal(85, 10, num_students), 1),
    'Parental_Education': np.random.choice(['High School', 'Graduate', 'Post Graduate'], size=num_students),
    'Sports_Participation': np.random.choice(['Yes', 'No'], size=num_students),
    'Math_Score': np.random.randint(30, 100, size=num_students),
    'Reading_Score': np.random.randint(35, 100, size=num_students),
    'Writing_Score': np.random.randint(40, 100, size=num_students),
}
df=pd.DataFrame(data)
df['Average_score']=df[['Math_Score', 'Reading_Score', 'Writing_Score']].mean(axis=1)   #here sir this .mean for average and axis 1 is ntg but column wise operation
df.head(5)      # why head means just for reference sir, I mean it will show firsst few records.
```

```python
print("Dataset Information in detail:")
print(df.info())
```

```python
print("\n Summary Statistics:")
print(df.describe())
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8,5))
se.scatterplot(x='Study_Hours', y='Average_score', hue='Gender', data=df, palette='Set2')  # so here the hue means its a seaborn related one, for coloring data points automatically. 
plt.title('Study Hours vs Average score')
plt.xlabel('Study Hours per Day')
plt.ylabel('Average score')
plt.show()
```

```python
plt.figure(figsize=(9,5))
se.scatterplot(x='Attendance_%', y='Average_score', hue='Gender', data=df, palette='coolwarm')
plt.title('Attendance vs Average score')
plt.xlabel('Attendance (%)')
plt.ylabel('Average score')
plt.show()
```

```python
plt.figure(figsize=(8,5))
se.boxplot(x='Parental_Education', y='Average_score', hue='Parental_Education', data=df, palette='pastel', legend=False)
plt.title('Average Score by Parental Education Level')
plt.xlabel('Parental Education')
plt.ylabel('Average Score')
plt.show()
```

```python
plt.figure(figsize=(6,5))
se.barplot(x='Gender', y='Average_score', hue='Gender', data=df, palette='muted', estimator='mean', legend=False)
plt.title('Average Performance by Gender')
plt.ylabel('Average Score')
plt.show()
```
